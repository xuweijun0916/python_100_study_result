```python
df = DataFrame(np.random.randint(0,150,size=(6,3)),index=list('ABCDEF'),columns=['python','math','en'])
df
```

```python
df3 = df.T
df3
```

```python
s = df.mean().round(1)
s
```

```python
df2 = DataFrame(s)
df2.columns = ['score_mean']
df2
```

```python
df3 = df2.T
df3
```

```python
df4 = df.append(df3)
df4
```

```python
df5 = DataFrame(df4.mean(axis=1).round(1))
df5.columns = ['score_mean']
df5
```

```python
df4.merge(df5,left_index=True,right_index=True)
```

```python
# 分组聚合透视
# 很多时候属性是相似的
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
df = DataFrame({'Hand':['right','left','left','right','right','right','right','right','left','right'],
               'Smoke':['yes','yes','no','no','yes','no','no','no','no','yes'],
               'sex':['male','female','female','male','male','male','female','female','male','female'],
               'weight':[80,50,48,75,68,100,40,90,88,76],
               'IQ':[100,120,90,130,140,80,94,110,100,160]})
df
```

```python
data = df.groupby(by=['Hand'])[['weight','IQ']].mean().round(1)
data
```

```python
df.groupby(by=['Hand'])[['weight']].apply(np.mean).round(1)
```

```python
data.agg(['max','mean']).round(1)
```
